## Diffusion Model — Inference Command

This example demonstrates **inference using a trained diffusion model**.

- **Model:** loaded from a saved checkpoint  
- **Inference year:** 2021  
- **Mode:** regional inference (Europe)

This configuration is intended for **evaluation and quick validation** of model outputs.

---

### Checkpoint Loading

You must specify the checkpoint file using:

- `--load_checkpoint_name`

This should be the **exact filename of the trained model checkpoint**, for example:  
`run_20260323_142528_final_model_epoch1.pth.tar`

The checkpoint must be located in:  
`checkpoints/notebook_test/test_run_diffusion`

---

### Inference Modes

Two inference modes are available:

- **Global inference**
  - `--run_type inference`
  - Runs on the full spatial domain  
  - No need to specify `region` or `region_size`

- **Regional inference**
  - `--run_type inference_regional`
  - Requires defining a spatial subset

#### Regional configuration

You can specify the region in two ways:

- **Predefined regions**
  - `--region europe`
  - `--region asia`
  - `--region us`

- **Custom region**
  - `--region_center lat lon`
  - `--region_size lat_size lon_size`

#### Region size

- `144 × 360` → **1 block**
- `288 × 720` → **4 blocks**

---

### Outputs

All outputs are saved in the following directories:

- **Results (plots, diagnostics):**  
  `results/notebook_test/test_run_diffusion`

- **Logs:**  
  `logs/notebook_test/test_run_diffusion`

```python
cmd = """
export PYTHONPATH="/leonardo_work/EUHPC_D27_095/kkingston/IPSL-AID:$PYTHONPATH"

cd "/leonardo_work/EUHPC_D27_095/kkingston/IPSL-AID"

python -m IPSL_AID.main \
    --debug false \
    --main_folder notebook_test \
    --sub_folder test_run_diffusion \
    --prefix run_20260323_142528 \
    --run_type inference_regional \
    --region Europe \
    --region_size 144 360 \
    --save_model true \
    --save_checkpoint_name difusion_model \
    --load_checkpoint_name run_20260323_142528_final_model_epoch1.pth.tar \
    --save_per_samples 10000 \
    --inference_type sampler \
    --arch adm \
    --precond edm \
    --in_channels 1 \
    --cond_channels 5 \
    --out_channels 1 \
    --apply_filter false \
    --model_name notebook_test_run_diffusion \
    --varnames_list VAR_2T \
    --normalization_types VAR_2T=standard \
    --constant_varnames_list z lsm \
    --constant_varnames_file ERA5_const_sfc_variables.nc \
    --units_list K \
    --year_start 2019 \
    --year_end 2019 \
    --year_start_test 2021 \
    --year_end_test 2021 \
    --batch_size 70 \
    --num_epochs 1 \
    --learning_rate 0.0001 \
    --num_workers 8 \
    --datadir /leonardo_work/EUHPC_D27_095/kkingston/AI-Downscaling/data/data_FOURxDaily \
    --per_var_datadir VAR_2T=/leonardo_work/EUHPC_D27_095/kkingston/AI-Downscaling/data/data_FOURxDaily \
    --time_normalization cos_sin \
    --tbatch 1800 \
    --sbatch 12 \
    --batch_size_lat 145 \
    --batch_size_lon 361 \
    --epsilon 0.02 \
    --beta 1.0 \
    --margin 8 \
    --dynamic_covariates_dir ../data_covariates/ \
    --dtype fp32 \
    --num_steps 20 \
    --sigma_min 0.002 \
    --sigma_max 80.0 \
    --rho 7 \
    --s_churn 40 \
    --solver heun \
    --compute_crps false
"""
```

```python
!bash -c "{cmd}"
```
